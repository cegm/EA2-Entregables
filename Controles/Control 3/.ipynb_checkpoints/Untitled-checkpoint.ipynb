```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

import sys
sys.path.insert(1, '/Users/cegm/Google Drive/ITAM/11/Estadística Aplicada II/Notas')
from Linear_Regression import Linear_Regression

pd.options.display.precision = 5 # Para  decimales.
plt.rcParams["font.family"] = "Helvetica Neue"
```

```python
df = pd.read_csv("bacterias.csv")
display(df.head())
```

```python
categorical = ["Temperatura [ºC]", "Cultivo", "Concentración"]
y = "Número final de bacterias"
```

```python
colors = {"Básico": "#3a86ff",
          "Ácido": "#ff006e"}

markers = {"Baja": "v",
           "Alta": "^"}

s = 300

big = 19
small = 17

plt.figure(figsize = (10, 10))
for cultivo, color in colors.items():
    for concentracion, marker in markers.items():
        matching_indices = (df["Cultivo"] == cultivo) & (df["Concentración"] == concentracion)
        
        for temperatura in df[categorical[0]].unique():
            print(f"Estadísticas para las observaciones con Temperatura de {temperatura} ºC, Cultivo {cultivo} y Concentración {concentracion}")
            subset_df = df.loc[matching_indices & (df[categorical[0]] == temperatura), y]
            print(f"{y} ∈ [{subset_df.min()}, {subset_df.max()}]")
        print("\n")
            
        plt.scatter(df.loc[matching_indices, categorical[0]], 
                         df.loc[matching_indices, y], 
                         c = color, 
                         marker = marker,
                         s = s, 
                         alpha = 0.55,
                         label = f"Cultivo {cultivo} y Concentración {concentracion}")
plt.tick_params(axis = 'both', which = 'major', labelsize = small)
plt.xlabel(categorical[0],fontsize=big)
plt.ylabel(y, fontsize=big)
plt.title(y + " vs. " + categorical[0].split(" [")[0],fontsize=big)
plt.legend(fontsize=small, title="Categorías", title_fontsize=big)
plt.show()
```

```python
# Obtener variables indicadoras para cada variable categórica

# Abreviar los nombres de las columnas categóricas
abbreviations = ["t", "k", "c"]

x = [] # Lista para almacenar los nombres de las variables indicadoras generadas

for category, abbreviation in zip(categorical, abbreviations):
    values = df[category].unique() # Valores distintos de la categoría actual
    n = len(values) # Cantidad de valores distintos de la categoría actual
    
    for k in range(n):
        value = values[k] # El valor actual de la categoría actual
        matching_rows = (df[category] == value) # Filas del DataFrame que son del valor actual de la categoría actual
        
        levels = f"{{:0{n-1}b}}".format(k)[::-1] # Código binario con los valores de cada variable indicadora
        
        for j in range(n-1): # Si la variable categórica tiene n valores se necesitan n-1 variables indicadoras
            label = fr"${{{abbreviation}}}_{j+1}$" if n > 2 else fr"${{{abbreviation}}}$" # Nombre de la variable indicadora
            x.append(label) # Agregar indicadora a la lista
            
            df.loc[matching_rows, label] = int(levels[j])

import collections
x = list(collections.OrderedDict.fromkeys(x)) # Quitar duplicados manteniendo el orden
```

```python
cod_temperatura = pd.read_csv("cod_temperatura.csv", index_col = 0)
cod_temperatura = cod_temperatura.style.set_caption("Codificación para las categorías del factor Temperatura [ºC]")
display(cod_temperatura)

cod_cultivo = pd.read_csv("cod_cultivo.csv", index_col = 0)
cod_cultivo = cod_cultivo.style.set_caption("Codificación para las categorías del factor Cultivo")
display(cod_cultivo)

cod_concentracion = pd.read_csv("cod_concentracion.csv", index_col = 0)
cod_concentracion = cod_concentracion.style.set_caption("Codificación para las categorías del factor Concentración")
display(cod_concentracion)

modelos = pd.read_csv("modelos.csv")
modelos = modelos.style.set_caption("Modelos propuestos para los distintos niveles de los factores")
pd.set_option('max_colwidth', 80)
pd.set_option('display.width', 1000)
display(modelos)
```

```python
# Obtener interacciones entre las variables indicadoras

import itertools
combinations = list(itertools.combinations(x, 2))

for combination in combinations:
    if combination != ('${t}_1$', '${t}_2$'): # Incluir esta combinación ocasiona singularidad
        label = f"Interacción {combination[0]} y {combination[1]}"
        x.append(label) # Agregar interacciones a la lista
    
        df[label] = df[combination[0]]*df[combination[1]]
```

```python
display(df)
for col in x: 
    print(col)
```

```python
intercept = True
alpha = 0.05

# x = x[0:5]

lr = Linear_Regression(df, x, y, intercept)
y_fit = f"Respuesta ajustada con {x} como regresor y {y} como respuesta"
df[y_fit] = lr.Y_fit
beta = lr.beta
print(beta)
# print(lr.to_string())

# scatter_fit_plot(df, peso, y, "#fc814a", y_fit_peso, lr_peso.to_string())
```
